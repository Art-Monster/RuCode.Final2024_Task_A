# Создание модели машинного обучения

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
# установим максимальный размер используемой оперативной памяти
h2o.init(max_mem_size='15G')
```

```python
# Загрузка данных
train_data = h2o.import_file("TaskA_cleaned_train_dataset_h2o_hope_v2.csv")
```

```python
# Разделение данных на обучающую и валидационную части
train_data, validation_data = train_data.split_frame(ratios=[0.8], seed=12)
```

```python
# Определение целевой переменной
target_column = "price"
```

```python
train_data.columns[:-1]
```

```python
# H2OAutoML?
```

```python
# Создание объекта AutoML
aml = H2OAutoML(
    balance_classes=True,
    max_models=80,  # Максимальное количество моделей для обучения
    max_runtime_secs=3600,  # Максимальное время обучения (в секундах)
    seed=12,  # Случайное число для воспроизводимости
    exclude_algos=["DeepLearning"],  # Исключить алгоритм Deep Learning
    nfolds=12,  # Количество фолдов для кросс-валидации
    stopping_metric="MAE",  # Метрика остановки
    stopping_tolerance=0.007,  # Допуск для остановки
    keep_cross_validation_predictions=True
)
```

```python
# Запуск AutoML
aml.train(x=['type', 'square', 'land_square', 
             'harmful_emissions_quant', 'mean_ppm', 'rooms_number_cleaned', 'bedrooms_number_cleaned', 'mean_room_size',
             'mean_ppm_land'], y=target_column, training_frame=train_data,
          validation_frame=validation_data  # Указываем валидационный датасет
         )
```

```python
# Получение лучшей модели
best_model = aml.get_best_model()
```

```python
# Вывод информации о лучшей модели
print(best_model)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
validation_data
```

```python
# Получение важности признаков
varimp = best_model.varimp()

# Вывод важности признаков
print(varimp)
```

```python
# Оценка модели
predictions = best_model.predict(validation_data)

# 2. Преобразуем прогнозы и реальные значения в numpy массивы
y_test = validation_data[target_column].as_data_frame().values.ravel()
y_pred = predictions.as_data_frame()["predict"].values.ravel()

mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"MAPE: {mape}")
```
